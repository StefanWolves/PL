```python
print(final_df)
```

```python
## creating min, max and average per position

average_df = final_df.groupby('Pos').mean()
min_df = final_df.groupby('Pos')['Pts'].min()
max_df = final_df.groupby('Pos')['Pts'].max()
```

```python
## Merging DataFrames

data_merged1 = pd.merge(left=average_df, right=min_df, how="left", left_on="Pos", right_on="Pos")
data_merged2 = pd.merge(left=data_merged1, right=max_df, how="left", left_on="Pos", right_on="Pos")
```

```python
## Renaming Headers and Converting Index to Column, sorting order of columns and dropping index

data_merged2.columns=['Won', 'Drew','Lost', 'Goal_Difference','Average_Points','Min_Points','Max_Points','Position']
data_merged2['Position'] = data_merged2.index
data_merged2.sort_values('Position')
columnstodrop = ['Pos']
data_merged3 = data_merged2.reset_index(drop=True)
data_merged4 = data_merged3[['Position', 'Won', 'Drew', 'Lost', 'Goal_Difference', 'Average_Points', 'Min_Points', 'Max_Points']]
print(data_merged4.to_string(index=True))
```

```python
## Plotting Average Points Per Position

data_merged4.sort_values('Average_Points').plot(x='Position', y='Average_Points', kind='barh')
```

```python
## Creating the standard deviation of min and max

standard_deviation = data_merged4["Max_Points"] - data_merged4["Min_Points"]
standard_deviation1 = standard_deviation.to_frame()
standard_deviation1.columns = ['Deviation'] 
```

```python
pd.merge(data_merged4, standard_deviation1, left_index=True, right_index=True)
```
